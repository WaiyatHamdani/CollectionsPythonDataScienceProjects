```python
import pandas as pd   #this to read csv using panda
import numpy as np
import matplotlib as plt
from sklearn import tree
```

```python
activitydata = pd.read_csv('../Data/csv/WaiyatActivity_oct4oct11.csv')
activitydata.head(10)
```

```python
date_step_data = activitydata [['Date','Steps']]
print(date_step_data)
```

```python
plt.pyplot.figure(figsize=(10, 5))
plt.pyplot.plot(activitydata['Date'], activitydata['Steps'], marker='o', linestyle='-.') # to make o shape marker and the line dot style is dash
plt.pyplot.xlabel('Date')
plt.pyplot.ylabel('Waiyat Step')
plt.pyplot.title('Waiyat Step/day Accoding samsung health')
plt.pyplot.xticks(rotation = 90)#to make the label vertical because it rotate 90 degree
plt.pyplot.show()
```

```python
# We going to reimport mathplotlib but with .pyplot to make mysely easier
import matplotlib.pyplot as p

p.figure(figsize=(12,6))
p.plot(activitydata['Date'],activitydata['Activity_Time_min'],marker = 'x', linestyle = '-')# to make x shape marker and the line style is dash
p.xlabel('Date')
p.ylabel('Waiyat minute active activity')
p.title('Waiyat Minute activity/day Accoding samsung health')
p.xticks(rotation = 45)#to make the label rotate 45 degree
p.show()
```

```python
p.figure(figsize=(15,6))
p.plot(activitydata['Date'],activitydata['Activity_Calories'],marker = 'x', linestyle = ':')# to make x marker and the line style is :
p.xlabel('Date')
p.ylabel('Waiyat active Calories')
p.title('Waiyat active Calories/day Accoding samsung health')
p.xticks(rotation = 35)#to make the label rotate 35 degree
p.show()
```

```python
p.figure(figsize=(15,6))
p.bar(activitydata['Date'],activitydata['Total_burned_Calories'])# to make it bar
p.xlabel('Date')
p.ylabel('Waiyat Total burned Calories')
p.title('Waiyat Total_burned_Calories/day Accoding samsung health')
p.xticks(rotation = 35)#to make the label rotate 35 degree
p.show()
```

```python
p.figure(figsize=(16,3))
p.hist(activitydata['Total_burned_Calories'], bins=4)
p.xlabel('Total burned')
p.ylabel('frequency')
p.title('Distribution of Waiyat total calorie burned per day')
p.xticks(rotation = 35)
p.show()
```

```python
#pandas filering data frame just for a certain date like 5,10,11
oct5_data=activitydata[activitydata['Date'] == 'Oct5'][['Steps', 'Activity_Calories', 'Total_burned_Calories']]
oct10_data=activitydata[activitydata['Date'] == 'Oct10'][['Steps', 'Activity_Calories', 'Total_burned_Calories']]
oct11_data = activitydata[activitydata['Date'] == 'Oct11'][['Steps', 'Activity_Calories', 'Total_burned_Calories']]
print(oct5_data)
print(oct10_data)
print(oct11_data)
```

```python
# using plot scatter for data oct 11 
p.scatter(oct5_data['Activity_Calories'],oct5_data['Total_burned_Calories'],color ='green' ,label='oct 5 Activity calories vs total calories')
p.scatter(oct10_data['Activity_Calories'],oct10_data['Total_burned_Calories'],color ='blue' ,label='oct 10 Activity calories vs total calories')
p.scatter(oct11_data['Activity_Calories'],oct11_data['Total_burned_Calories'],color ='red' ,label='0ct 11 Activity calories vs total calories')

p.xlabel('Activity_Calories')
p.ylabel('Total Calories burned')
p.title('Activity_Calories vs Total_Calories_Burned by Waiyat')
p.legend() # to show the label title 
p.show()
```

```python
#okay im going to do for loop for the rest of scatter data
unique_dates = activitydata['Date'].unique()

for date in unique_dates:
    # all data in the current date after for looping them 
    date_data = activitydata[activitydata['Date']==date]
    #then we can scatter them 
    p.scatter(date_data['Activity_Calories'], date_data['Total_burned_Calories'], label=f'{date} Activity vs Total')
    
p.xlabel('Activity_Calories')
p.ylabel('Total Calories burned')
p.title('Activity_Calories vs Total_Calories_Burned by Waiyat')
p.legend() # to show the label title 
p.show()
```

```python
# we going to trace the line from beginning oct 5 - oct 11 with each point different color

# Define unique colors for each point
colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'cyan', 'magenta']

# Plot the continuous line connecting all the points
p.plot(activitydata['Activity_Calories'], activitydata['Total_burned_Calories'], 
         linestyle='-', color='gray', label='All Days Combined')

# Plot individual points with different colors
# loop Enumerate explanation -> for i, date in enumerate(activitydata['Date']):
# ['Oct4', 'Oct5', 'Oct6', 'Oct7', 'Oct8', 'Oct9', 'Oct10', 'Oct11'] it provide date and the index
# i = 0 -> it gave value : 'Oct4'
for i, date in enumerate(activitydata['Date']):
    p.scatter(activitydata['Activity_Calories'][i], activitydata['Total_burned_Calories'][i], 
                color=colors[i % len(colors)], label=f'{date} Activity vs Total')

# Add labels and title
p.xlabel('Activity Calories')
p.ylabel('Total Calories Burned')
p.title('Activity Calories vs Total Calories Burned by Waiyat (Different Colors for Each Day)')
p.legend()# to show the label title 
p.show()
```
